###  Subscripts

Las funciones operan sobre los vectores como un todo. Sin embargo hay ocasiones en las que solo se requiere aplicar las funciones a determinados elementos de un vector. 
Esto se consigue con los subscripts. Los subscripts están delimitados por corchetes [], y las funciones por paréntesis ().

```r
x <- 0:10
x
```

```r
sum(x) # Suma de todos los elementos de x
```

```r
sum(x<7) # Número de elmentos de x que cumplen la restricción x<5
```

```r
sum(x[x<5]) # Suma de los elementos que cumplen con la restricción x<5
```

```r
y <- c(5,8,6,7,1,5,3)
y
```

```r
y[4] # Despliga el 4to elemento del objeto "y"
```

```r
z <- y[-1] # Se crea el objeto z con todos los elementos de "y" menos el primero
z  
```

Supongamos que queremos una función que calcule la media de un vector que excluya el valor máximo y mínimo


```r
w <- c(12,4,5,3,7,8,6,1,11,7)
```

```r
(3+4+5+6+7+7+8+11)/8
```

```r
w=sort(w)
```

```r
w
```

```r
w=w[-1] # Enlista los elementos de w excluyendo el primero (1)
```

```r
w
```

```r
length(w) # Número de elementos de w
```

```r
w=w[-length(w)] # Enlista los elementos de w excluyendo el último (12)
```

```r
wp= sum(w)/length(w)
```

```r
length(w)
```

```r
wp
```

```r
-c(1,length(w))# Elementos que habremos de excluir
```

```r
sort(x)[-c(1,length(w))] # Elementos ordenados de w donde se excluye el primero y el décimo
```

```r
mean(sort(w)[-c(1,length(w))]) # Promedio de los elementos ordenados de w excluyendo el mínimo y el máximo
```

```r
promedio <- function (w) mean(sort(w)[-c(1,length(w))])  
```
