### Gráficas con los datos de pDF

```r
plot(mujeres4560$escoacum, mujeres4560$numhij, xlab="Años de escolaridad", ylab="Número de hijos")
```

```r
# Significancia de la asociación

cor.test(mujeres4560$escoacum, mujeres4560$numhij, use="complete.obs")
```
