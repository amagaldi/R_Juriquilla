Crear una variable usando una ciclo if/else

Identificar a las solteras con hijos (sch)

```r
table(pDF$numhij, pDF$estcon) 
```

```r
for (i in 1:nrow(pDF)) {
  
  if (pDF$estcon[i] == 8 & pDF$numhij[i] != 0 & pDF$numhij[i] != 99) {
    pDF$sch[i] <- 1 # Solteras con hijos
  } else {if (pDF$sexo[i] == 1) {
    pDF$sch[i] <- 3 # Hombres
  } else {pDF$sch[i] <- 2} # Solteras sin hijos
  }}
```

```r
table(pDF$sch)
```

Generar un factor a partir de la variable que acabamos de crear

```r
pDF$sch_fac <- factor(pDF$sch, levels=1:3, 
                      labels=c("Sí", "No", "Hombre"))
```

```r
table(pDF$sch_fac)
```

 Pegado de base de datos

```r
pDF$id <- c(1:nrow(pDF))
```

```r
head(pDF)
```

```r
base1 <- pDF[,c("id", "edad")]
```

```r
base2 <- pDF[,c("id", "parent")]
```

```r
head(base1)
```

```r
head(base2)
```

Modo simple de pegar (pero peligroso): cbind  

```r
base3 <- cbind(base1, base2)
```

```r
head(base3)
```

Precaución! Los archivos pueden estar desordenados u ordenados de forma diferente!


```r
base2_decr <- base2[order(base2$id, decreasing=TRUE), ]
```

```r
base4 <- cbind(base1, base2_decr)
```

```r
head(base4)
```

Forma más segura: usar función merge(). 

```r
base5 <- merge(base1, base2_decr, by="id", all=TRUE)  
```

Con 'all=TRUE'se mantienen todos los casos, con 'all=FALSE' 
se excluyen aquellos que no tienen el mismo ID
